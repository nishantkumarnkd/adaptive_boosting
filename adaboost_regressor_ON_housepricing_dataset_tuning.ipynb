```python
import numpy as np
import pandas as pd
house = pd.read_csv('housebuilding.csv')
```

```python
#threshold is set at 1430 means to retain only those columns which has at least 1430 non NAs from 1460 total values
house.dropna(axis='columns',thresh=1430,inplace=True) 
house.isnull().sum()
#to fill columns with few missing values
house.fillna({'MasVnrType':'BrkFace','MasVnrArea':np.mean(house['MasVnrArea'])},inplace=True)
house.fillna('SBrkr',inplace=True)
house.isnull().sum()
```

```python
# splitting our dataset into train data and test data
from sklearn.model_selection import train_test_split
y = house['SalePrice']
house.drop(columns='SalePrice',inplace=True)
X_train,X_test,y_train,y_test = train_test_split(house,y,test_size=0.2,random_state=42)
```

```python
#label encoding our columns which are categorical in nature
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in X_train.columns.values:
    if X_train[col].dtypes == 'object':
        data = X_train[col].append(X_train[col])
        le.fit(data.values)
        X_train[col]= le.transform(X_train[col])
X_train
```

```python
#label encoding the categorical columns present in our test data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in X_test.columns.values:

       if X_test[col].dtypes=='object':
            data=X_test[col].append(X_test[col])
            le.fit(data.values)
            X_test[col]=le.transform(X_test[col])
X_test
```

```python
#applying adaptive boosting without any HYPERPARAMETRIC TUNING 
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()
ada.fit(X_train,y_train)
predicted = ada.predict(X_test)
predicted
```

```python
#checking our r square score
from sklearn.metrics import r2_score
r2_score(y_test,predicted)
```

```python
"""now we are doing Hyperparametric Tuning to increase our model's accuracy score
by using RandomisedSearchCV"""

from sklearn.model_selection import RandomizedSearchCV
import numpy as np
random_tuning = {'n_estimators':range(40,130,10),
                 'learning_rate':np.arange(0.1,2,0.2)}
random = RandomizedSearchCV(estimator=ada,n_iter=100,cv=5,param_distributions=random_tuning,random_state=42)
random.fit(X_train,y_train)
random.best_params_
#here we get 110 as no.of trees we should give to our model and the learning rate as 0.3
```

```python
"""Hyperparametric Tuning to increase our model's accuracy score
by using GridSearchCV on the previous parameters we got from randomisedsearchcv"""

from sklearn.model_selection import GridSearchCV
tuning2 =  { 'n_estimators':[100,110,120],
          'learning_rate':[0.2,0.3,0.4]
          }
grid = GridSearchCV(estimator=ada,param_grid=tuning2,cv=5)
grid.fit(X_train,y_train)
grid.best_params_
```

```python
#again applying adaboost with hyperparameters and we can see an improvement in our score
from sklearn.ensemble import AdaBoostRegressor
best_adaptive = AdaBoostRegressor(n_estimators=120,learning_rate=.3)
best_adaptive.fit(X_train,y_train)
best_predicted = best_adaptive.predict(X_test)
best_predicted
```

```python
#checking score again after hyperparametric tuning
r2_score(y_test,best_predicted)
```
